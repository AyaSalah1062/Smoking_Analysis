* (1) indicates a strong positive correlation: as one variable increases, the other variable tends to increase as well.
-(-1) indicates a strong negative correlation: as one variable increases, the other variable tends to decrease.

**Skewed Right (Positive Skewness):**

* Description: A distribution is said to be right-skewed when the right tail (larger values) is longer or fatter than the left tail.
* Implications:
Most of the values are concentrated on the left side.
The mean is typically greater than the median.
* The distribution may have potential outliers on the right side.

* Right-skewed distributions are common and may not necessarily be "bad." It indicates that there might be some extreme values on the higher end.

**Outliers:**

* Definition: Outliers are data points that deviate significantly from the rest of the data.
* Visibility:
Outliers are often visible as individual points or dots outside the main body of the distribution on plots like box plots.
* Implications:
Outliers can impact statistical analyses and machine learning models.

* Good: Outliers may contain valuable information, especially in certain research or anomaly detection scenarios.
* Bad: Outliers can also introduce noise and affect the assumptions of some statistical tests.
